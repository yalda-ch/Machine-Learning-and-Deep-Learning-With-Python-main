```python
# Visualization of a Sine Curve with Noise-Corrupted Data Points
import matplotlib.pyplot as plt
import numpy as np
x_true = np.arange(0, 1 ,0.01)
y_true = np.sin(2 * np.pi * x_true)
x_sample = np.arange(0, 1, 0.1)
y_sample = [0.125,0.5,0.98,0.92,0.625,-0.1,-0.7,-0.92,-0.98,-0.68] #Manually added noise to simulate observed data
#y_sample = np.sin(2 * np.pi * x_sample) + 0.1 * np.random.randn(10) Add random Gaussian noise to the data
plt.plot(x_true, y_true, label="true underlying function")
plt.plot(x_sample, y_sample, 'bo', label="observed data")
plt.title("Plot of noise corrupted data")
plt.axis([-0.05, 1, -1.1, 1.1])
plt.legend()
plt.show()
```

```python
#Check Whether an Array is Monotonic
def is_monotonic(array):
  up = 0 #increasing
  down = 0 #decreasing
  for i in range(1, len(array)):
    if array[i] > array[i-1]:
      up += 1
    elif array[i] < array[i-1]:
      down += 1
  if up > 0 and down > 0:
      print(array,"is not monotonic.") #both increasing and decreasing.
  else:
      print(array,"is monotonic.") #only an increase or only a decrease.
is_monotonic([1, 2, 3, 3, 3, 4, 5, 6, 7])       # monotonic
is_monotonic([7, 6, 5, 4, 3, 3, 3, 2, 1])      # monotonic
is_monotonic([1, 5, 4, 3, 3, 3, 2, 6, 7])  # not monotonic
```
